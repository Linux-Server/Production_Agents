```python
## What sre the thing get stored in a vector database? 
# Vectors [,231,3,31e,4,134e]
# Text
## INDEXING
from langchain_community.document_loaders import WebBaseLoader
import bs4

parse_data = bs4.SoupStrainer(class_=("post-title", "post-header", "post-content"))

docs = WebBaseLoader(web_path="https://lilianweng.github.io/posts/2023-06-23-agent/", bs_kwargs={"parse_only": parse_data})

docs = docs.load()
```

```python
print(docs[0].page_content)
```

```python
## Split the data
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    add_start_index=True
)
```

```python
all_doc_split = splitter.split_documents(documents=docs)
```

```python
all_doc_split[0]
```

```python
## Embed ans vector store
from langchain_openai.embeddings import OpenAIEmbeddings
from langchain_community.vectorstores import InMemoryVectorStore

embedding = OpenAIEmbeddings()

store = InMemoryVectorStore(embedding=embedding)
```

```python
store_ids = store.add_documents(all_doc_split)
store_ids
```

```python
## retrival
from langchain.agents import create_agent
from langchain.tools import tool

@tool
def retreive_context(query:str):
    """Retreive the context based on query"""
    contexts = store.similarity_search(query, k=2)
    res = " ".join([doc.page_content for doc in contexts])  #
    return res
```

```python
agent = create_agent(
    model="gpt-5-nano",
    tools=[retreive_context],
    system_prompt=(
    "You have access to a tool that retrieves context from a blog post. "
    "Use the tool to help answer user queries."
)
)
```

```python
agent.invoke({"messages":[{"role":"user", "content" :"What is Short-term memory and Self reflection?"}]})
```

```python
for event in agent.stream(
    {"messages": [{"role": "user", "content": "What is Short-term memory and GPT-Engineer Project?"}]},
    stream_mode="values",
):
    event["messages"][-1].pretty_print()
```

```python
## twp step RAG chain
from langchain.agents.middleware import dynamic_prompt, ModelRequest

@dynamic_prompt
def prompt_with_context(request:ModelRequest):
    """Inject context"""
    print(request.state["messages"][-1].content)
    system_prompt = "your are an helpful AI assistant"
    return system_prompt


agent = create_agent(
    model="gpt-5-nano",
    middleware=[prompt_with_context]
)
```
