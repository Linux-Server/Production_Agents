```python
## Load the doc
with open("text.txt") as f:
    contents = f.read()
```

```python
from rich.markdown import Markdown

Markdown(contents)
```

```python
## split documents
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=256,
    chunk_overlap=25
)
```

```python
data = splitter.split_text(contents)
```

```python
len(data)
```

```python
## Embedding
from langchain_community.embeddings import OpenAIEmbeddings

embedings = OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
## vector store
from langchain_community.vectorstores import InMemoryVectorStore

store = InMemoryVectorStore(embedding=embedings)
```

```python
database = store.add_texts(data)
```

```python
query = "Tell me population of kerala"

store.similarity_search(query, k=2)
```

```python
from langchain.agents import create_agent
from langchain.tools import tool

@tool
def retrieve_context(query:str):
    """Retreive the data from vector store based on query"""
    out = store.similarity_search(query=query, k=2)
    res =  " ".join([doc.page_content for doc in out])
    print("Res", res)
    return res


agent = create_agent(
    model="gpt-5-nano",
    tools=[retrieve_context],
    system_prompt="If the user ask about Kerala, use the tool to retreive the context"

)
```

```python
agent.invoke({"messages": [{"role": "user", "content": "Tell me about kerala"}]})
```
