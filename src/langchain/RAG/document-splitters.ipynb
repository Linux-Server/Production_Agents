### Documents Text Splitters

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=50,
    chunk_overlap=5
    
)

### First prioti --> sentence , para, new line, space. --> split
```

```python
text = """Qwen3 represents a significant milestone in our journey toward Artificial General Intelligence (AGI) and Artificial 
Superintelligence (ASI). By scaling up both pretraining and reinforcement learning (RL), we have achieved higher   
levels of intelligence."""

chunks = splitter.split_text(text)

chunks

```

```python
len(chunks[4])
```

```python
# Show size + content
for i, chunk in enumerate(chunks, 1):
    print(f"Chunk {i:2d}  |  {len(chunk):2d} chars  →  {chunk!r}")
```
